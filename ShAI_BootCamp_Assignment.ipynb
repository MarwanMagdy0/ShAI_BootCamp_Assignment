### Task3

```python
df["BasePay"] = df["BasePay"].fillna(df["BasePay"].mean())
df["Benefits"] = df["Benefits"].fillna(df["Benefits"].mean())
df["OvertimePay"] = df["OvertimePay"].fillna(df["OvertimePay"].mean())
df["OtherPay"] = df["OtherPay"].fillna(df["OtherPay"].mean())
df.drop(columns=["Notes", "Status"], inplace=True)
```

### Task4

```python
import matplotlib.pyplot as plt

# Assuming 'TotalPay' is the column representing the total pay of employees
plt.hist(df['TotalPay'], bins=10, edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['JobTitle'][:50].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Agencies')
plt.ylabel('') 
plt.show()
```

```python
average_salaries = df.groupby('Agency')['TotalPay'].mean()
print(average_salaries)
```

### Task5

```python
df["JobTitle"].unique()
```

```python
average_totalpay = df.groupby('JobTitle')['TotalPay'].mean().sort_values()
print(average_totalpay)
```

### Task6

```python
correlation = df['TotalPay'].corr(df['BasePay'])
print(correlation)
```

```python
import seaborn as sns
correlation_matrix = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay']].corr()

# Plot the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['TotalPay'], df['BasePay'])
plt.title('Scatter plot of TotalPay vs BasePay')
plt.xlabel('TotalPay')
plt.ylabel('BasePay')
plt.show()
```
